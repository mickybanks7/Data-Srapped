```python
import re
import requests
from lxml import html
from bs4 import BeautifulSoup as bs

url ='http://www.nairaland.com/home'
html = requests.get(url) 

#print (html.text)

data = html.text

soup_data = bs(data, 'lxml') 

soup_data("td")
```

```python
for data in soup_data("td"):
    print (data.text)
```

```python
member_exist = None

data_match = "[\w]+\([\d]+\)"

for data in soup_data("td"):
    data_found = re.findall(data_match, data.text)
    
    if data_found:
        member_exist = data_found

print (member_exist)
```

```python
member_exist_replaced = [x.replace(")", "") for x in member_exist]          

print (member_exist_replaced)
```

```python
for e in member_exist_replaced:
    member_cleaned = e.split("(")
    print (member_cleaned)
    
```

```python
member_cleaned = {}

for y in member_exist_replaced:
    temp_data = y.split("(")
    
    member_cleaned[temp_data[0]] = int(temp_data[1])
    
print (member_cleaned)
```

```python
import pandas as pd
from datetime import datetime
import seaborn as sns
%matplotlib inline

data_f = pd.DataFrame(list(member_cleaned.items()), columns = ["Username", "Age"] )

data_f
```

```python
todays_date = datetime.now().date()

data_f["Date"] = todays_date

data_f
```

```python
csv_name = todays_date.strftime("%Y%m%d")
data_f.to_csv(csv_name + ".csv")
```
